```python
import matplotlib.pyplot as plt
```

```python
def fx1(x2):
    return (8-x2)/3.0

def fx2(x1):
    return (20-5.0*x1)/5.0

def jacobi(initialguess,iter,tol):
    x1_data = []
    x2_data = []

    for _ in range(iter):
        initialvalue = (initialguess[0],initialguess[1])
        x1 = fx1(initialguess[1])
        x2 = fx2(initialguess[0])
        initialguess[0] = x1
        initialguess[1] = x2
        x1_data.append(x1)
        x2_data.append(x2)
        if abs(initialvalue[0]-initialguess[0])<tol and abs(initialvalue[1]-initialguess[1])<tol:
            break
    
    return [x1_data,x2_data]

def seidal(initialguess,iter,tol):
    x1_data = []
    x2_data = []
    for _ in range(iter):
        initialvalue = (initialguess[0],initialguess[1])
        initialguess[0]=fx1(initialguess[1])
        initialguess[1]=fx2(initialguess[0])
        x1_data.append(initialguess[0])
        x2_data.append(initialguess[1])
        if abs(initialvalue[0]-initialguess[0])<tol and abs(initialvalue[1]-initialguess[1])<tol:
            break
    
    return [x1_data,x2_data]

tol = 1e-6
initialguess1 = [0,0]
initialguess2 = [0,0]
data_jacobi = jacobi(initialguess1,100,tol)
data_seidal = seidal(initialguess2,100,tol)
```

```python
plt.scatter(data_jacobi[0],data_jacobi[1])
plt.xlabel('x1')
plt.ylabel('x2')
plt.title("Gauss-Jacobi")
plt.show()
```

```python
plt.scatter(data_seidal[0],data_seidal[1])
plt.xlabel('x1')
plt.xlabel('x1')
plt.title("Gauss-Seidal")
plt.show()
```
